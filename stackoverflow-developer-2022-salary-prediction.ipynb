```python
def plot_line_chart(df, column, line=""):
    if line == '':
        line = df[column].value_counts().keys()[0:20]
    data = df[column].value_counts()[0:20]
    fig=px.line(x=line,y=data)
    fig.show()
    
def plot_bar_chart(df, column, line=""):
    if line == '':
        line = df[column].value_counts().keys()[0:20]
    data = df[column].value_counts()[0:20]
    fig=px.bar(x=line,y=data)
    fig.show()
    
def plot_pie_chart(df, column, line = ''):
    if line == '':
        line = df[column].value_counts().keys()[0:20]
    data = df[column].value_counts()[0:20]
    fig=px.pie(names=line,values=data)
    fig.show()
```

```python
df = df[["Country", "EdLevel", "YearsCodePro","Employment", "ConvertedCompYearly"]]
df = df.rename({"ConvertedCompYearly": "Salary"}, axis=1)
df.head()
```

```python
def print_unique_col_values(df):
       for column in df:
            if df[column].dtypes=='object':
                print(f'{column}: {df[column].unique()}') 
```

```python
print_unique_col_values(df)
```

```python
df = df[df["Employment"] == "Employed, full-time"]
df = df.drop("Employment", axis=1)
df.info()
```

```python
df['Country'].value_counts()
```

```python
def shorten_categories(categories, cutoff):
    categorical_map = {}
    for i in range(len(categories)):
        if categories.values[i] >= cutoff:
            categorical_map[categories.index[i]] = categories.index[i]
        else:
            categorical_map[categories.index[i]] = 'Other'
    return categorical_map
```

```python
country_map = shorten_categories(df.Country.value_counts(), 199)
df['Country'] = df['Country'].map(country_map)
df.Country.value_counts()
```

```python
fig = px.box(df, x="Country", y="Salary")
fig.show()
```

```python
df = df[df["Salary"] <= 300000]
df = df[df["Salary"] >= 10000]
df = df[df['Country'] != 'Other']
```

```python
fig = px.box(df, x="Country", y="Salary")
fig.show()
```

```python
df["YearsCodePro"].unique()
```

```python
def clean_experience(x):
    if x ==  'More than 50 years':
        return 50
    if x == 'Less than 1 year':
        return 0.5
    return float(x)

df['YearsCodePro'] = df['YearsCodePro'].apply(clean_experience)
```

```python
df["YearsCodePro"].unique()
```

```python
def clean_education(x):
    if 'Bachelor’s degree' in x:
        return 'Bachelor’s degree'
    if 'Master’s degree' in x:
        return 'Master’s degree'
    if 'Professional degree' in x or 'Other doctoral' in x:
        return 'Post grad'
    return 'Less than a Bachelors'

df['EdLevel'] = df['EdLevel'].apply(clean_education)
```

```python
df["EdLevel"].unique()
```

```python
 X = df.drop("Salary", axis=1)
 y = df["Salary"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.20, random_state=100)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.compose import ColumnTransformer
```

```python
categorical_pipeline = Pipeline([('OneHot', OneHotEncoder())])


numeric_pipeline = Pipeline([('Scaler', StandardScaler())]) 
 
                                 
transformer = ColumnTransformer([('Num', numeric_pipeline, ['YearsCodePro']), 
                                 ('Category', categorical_pipeline, ['Country','EdLevel'])])
```

```python
mlpipe1 =Pipeline([('transformer',transformer),('lr',LinearRegression())])
```

```python
mlpipe1.fit(X_train,y_train)
```

```python
y_hat = mlpipe1.predict(X_test)
y_hat
```

```python
error = np.sqrt(mean_squared_error(y_test,y_hat))
print ('${:.02f}'.format(error))
```

```python
mlpipe2 =Pipeline([('transformer',transformer),('xgb',XGBRegressor())])
mlpipe2.fit(X_train,y_train)
y_hat = mlpipe2.predict(X_test)
y_hat
error = np.sqrt(mean_squared_error(y_test,y_hat))
print ('${:.02f}'.format(error))
```
